```
import shutil
import os
import json
import objectpath
entire_data_set = []
data_set = 'biorxiv_medrxiv/'
for r, d, f in os.walk(data_set):
    for file in f:
      filename = os.path.join(r, file)
      with open(filename) as datafile:
        data = json.load(datafile)
        jsonnn_tree = objectpath.Tree(data)
        paper_id = tuple(jsonnn_tree.execute('$..paper_id'))
        text = tuple(jsonnn_tree.execute('$..text'))
        paper_and_text = [paper_id,text]
        entire_data_set.append(paper_and_text)
```

```
s = 'Paper ID :' + str(entire_data_set[0][0]) + ' ' + 'Content :' +  str(entire_data_set[0][1])
```

```
content = s.replace('\xa0', ' ').replace('\n', ' ').replace('(', ' ').replace(')', ' ').replace('{', ' ').replace('}', ' ').replace(':', ' ').replace('"', ' ').replace('"', ' ')
```

```
content = content[0:255]
```

```
question = 'How does geography affect Coronavirus virality ?'
```

```
input_data = content + question
```

```
gpt2_input_representation = '/content/gpt-2qa/input_file.txt' 
```

```
with open(gpt2_input_representation, "w") as f1:
  f1.write(input_data)
```
